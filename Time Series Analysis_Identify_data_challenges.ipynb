```python
#lets fix the order date as the index and use the reindex funtion to fill 0 values in the missing dates
data_date_idx = data.set_index('Order Date')
idx = pd.date_range(data_date_idx.index.min(), data_date_idx.index.max())


idx
```

```python
len(idx)
```

```python
#Simply reindexing won't work as the order date column has duplicates
data_date_idx.index.duplicated()
```

```python
#a slightly convoluted way of adding in the missing date values
new_df = data.groupby('Order Date',as_index=False).sum()[['Order Date','Sales']]

missing = idx.difference(new_df['Order Date'])

for date_idx in missing:
  new_row = {'Order Date': date_idx,'Sales':0}
  new_df = new_df.append(new_row,ignore_index=True)

new_df

```
