```python
#Lets get the text document we'll be analysing. Its a Harry Potter fanfiction
#Some new words might be out of our dictionary

f = open("text.txt","r",encoding="utf-8")
text = f.read()

text
```

```python
import spacy

# Load the large English NLP model
nlp = spacy.load('en_core_web_lg')

# Parse the text with spaCy. This runs the entire pipeline.
# so we've already segmented and tokenized the doc in one line
doc = nlp(text)

#with this pipeline we can look at different kind of words and phrases
#for instance let us look at flat phrases that have a noun as their head
for chunk in doc.noun_chunks:
    print(chunk.text, chunk.root.text, chunk.root.dep_,
            chunk.root.head.text)
```

```python
from spacy import displacy

sents=text.split('.')

sents[1]
```

```python
sent_token= nlp(sents[1])

displacy.render(sent_token, style='dep', jupyter=True, options={'distance':100})
```

```python
# Let us tag our tokens and identify what part of speech they belong to
print("\n\nPoS tagging\n")
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)

```

```python
#Text Lemmatization: finding the base word of our word in doc
# for instance stand becomes standing, fell becomes fall
# we'll put lemmatized words in a new filtered doc
filter_doc=""
for token in doc:
  filter_doc+=token.lemma_ + " "
  if token.text != token.lemma_:
    print(token.text+" becomes "+token.lemma_)
```

```python
filter_doc
```

```python
# now we'll filter out stop words
doc = nlp(filter_doc)
filter_doc=""
for token in doc:
  if token.is_stop:
    print(token.text)
  else:
    filter_doc+=token.text+ " "
```

```python
#lets check how spacy tags the words in our doc
doc = nlp(filter_doc)
for entity in doc.ents:
    print(f"{entity.text} ({entity.label_})")
```

It tags Hufflepuff as person :)

```python
import textacy


uniqueStatements = []
for word in ["Tabitha Longbottom", "Tabitha", "Longbottom"]:
    for cue in ["be", "have", "write", "talk", "talk about"]:
        statments = textacy.extract.semistructured_statements(doc, entity=word, cue = cue )
        for statement in statments:
            uniqueStatements.append(statement)

print("found", len(uniqueStatements), "statements.")
for statement in uniqueStatements:
    entity, cue, fact = statement
    print("* entity:",entity, ", cue:", cue, ", fact:", fact)
```
